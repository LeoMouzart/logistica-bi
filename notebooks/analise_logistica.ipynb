### Tratando Valores Nulos 

Nesse espaço realizarei as tratativas referentes a valores nulos na tabela entendendo o que cada um deles representa.

```python
df.columns
```

##### COLUNA TRANSPORTADORA 

```python
df["Transportadora"] = df["Transportadora"].fillna("Não Informado")
```

```python
df["Transportadora"].isnull().sum()
```

##### COLUNA MODALIDADE

```python
df["Modalidade"] = df.groupby("Categoria_Produto")["Modalidade"].transform(
    lambda x: x.fillna(x.mode()[0] if not x.mode().empty else "Rodoviario")
)
```

```python
df["Modalidade"].isnull().sum()
```

#### COLUNAS NÚMERICAS 

<p>Aqui a tratativa não sera um valor fixo, preencherei com o calculo da mediana justamente por conta do desvio padrão alto


#### Peso e Volume

```python
df["Peso_KG"] = df["Peso_KG"].fillna(df["Peso_KG"].median())
df["Volume_M3"] = df["Volume_M3"].fillna(df["Volume_M3"].median())
```

```python
df["Peso_KG"].isnull().sum()
df["Volume_M3"].isnull().sum()
```

#### Dias Reais de Entrega

```python
df["Dias_Reais_Entrega"] = df.groupby("Status_Entrega")["Dias_Reais_Entrega"].transform(
    lambda x:x.fillna(x.median())
)
```

```python
df["Dias_Reais_Entrega"].isnull().sum()
```

#### Custo Frete BRL

```python
df["Custo_Frete_BRL"] = df["Custo_Frete_BRL"].fillna(df["Custo_Frete_BRL"].median())
```

#### Satisfacao do CLiente

```python
df["Satisfacao_Cliente"] = df["Satisfacao_Cliente"].fillna(df["Satisfacao_Cliente"].median())
```

```python
df["Satisfacao_Cliente"].isnull().sum()
```

#### Confirmando se todas as colunas estão com os dados preenchidos

```python
df.isnull().sum()
```

#### Criando Coluna para análise de performance com base na entrega

```python
df["Desvio_Dias"] = df["Dias_Reais_Entrega"] - df["Prazo_Contratado_Dias"]
```

```python
df["Desvio_Dias"].describe().round(2)
```

```python
df["Pedido_Atrasado"] = (df["Desvio_Dias"] > 0).astype(int)
df["Pedido_Atrasado"].value_counts()
```

```python

(df["Pedido_Atrasado"].mean().round(2))
```

```python
df["Custo_Por_KG"] = (df["Custo_Frete_BRL"] / df["Peso_KG"]).round(2)
```

```python
df["Mês"] = df["Data_Pedido"].dt.month
df["Trimestre"] = df["Data_Pedido"].dt.quarter
df["Ano"] = df["Data_Pedido"].dt.year
df["Mês_Ano"] = df["Data_Pedido"].dt.to_period("M").astype(str)
```

### 💡 Insight 3 — Prejuízo com Extravios
R$ 328.633,36 em mercadorias extraviadas + R$ 22.924,48 em

Categoria Alimentos concentra R$ 72.167,34 desse total —

### 💡 Insight 4 — Satisfação x Atraso
Pedidos entregues no prazo têm satisfação média de 3.09
contra 2.90 dos entregues com atraso.

Clientes em status intermediários (Aguardando Coleta, Em Trânsito)
avaliam melhor por ainda terem expectativa positiva.

**Recomendação:** comunicar proativamente atrasos ao cliente —
a surpresa negativa impacta mais do que o atraso em si.

```python
financeiro = df.groupby("Transportadora").agg(
    Custo_Total  = ("Custo_Frete_BRL", "sum"),
    Valor_NF     = ("Valor_NF_BRL", "sum"),
).round(2)

financeiro["Pct_Frete_NF"] = (financeiro["Custo_Total"] / financeiro["Valor_NF"] * 100).round(2)

financeiro.sort_values("Custo_Total", ascending=False)
```

### 💡 Insight 5 — Financeiro
Azul Cargo transporta o maior valor em mercadorias (R$ 348k)

transportadora com maior taxa de atraso e extravios.

Latam Cargo tem o maior custo relativo (6.48%) mas melhor
satisfação — o custo extra pode ser justificado pela qualidade.

**Recomendação:** redistribuir cargas de alto valor da Azul Cargo
para transportadoras com menor índice de extravio.
